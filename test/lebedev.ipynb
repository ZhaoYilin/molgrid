```python
import numpy as np
import pandas as pd

# 加载数据
data = np.load('../grid/data/lebedev_131_5810.npz')
data2 = np.loadtxt('../grid/data/lebedev_131.txt')


angles = data2[:,0:2]            
            
# 提取数据
points = data['points']
weights = data['weights'].flatten()  # 确保权重是一维数组

# 方法2b：使用pandas（推荐，更易读）
df_coords = pd.DataFrame(points, columns=['x', 'y', 'z'])
df_angles = pd.DataFrame(angles, columns=['theta', 'phi'])  # 如果有角度数据
df_angles['r'] = 1
df_angles = df_angles[['r', 'theta', 'phi']]

# 方法1a：使用concat水平合并
df_combined = pd.concat([df_angles, df_coords], axis=1)
df_combined['weight'] = weights.astype(np.float64)


df_combined = df_combined.astype(np.float64)


df_combined.to_csv('../grid/data/lebedev_131_5810.csv', index=False)

# 显示前几行确认
print("\n数据预览:")
print(df_combined.head())
```
