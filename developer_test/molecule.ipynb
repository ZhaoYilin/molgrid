```python
# Create water molecule
atoms = [
    Atom('O', [0.0, 0.0, 0.0]),
    Atom('H', [0.757, 0.586, 0.0]),
    Atom('H', [-0.757, 0.586, 0.0])
]
    
mol = Molecule(atoms, charge=0, multiplicity=1)
    
print(mol.atoms)
print(mol.charge, mol.multiplicity, mol.mass)

for atom in mol:
    print(f"{atom.symbol}: {atom.cov_radius_slater:.2f} Å")
    
```
